```python
import pandas as pd
import os as os
import seaborn as sns
import matplotlib.pyplot as plt



df = pd.read_csv("peliculas.csv",sep=",",engine="python")
print(df)
grafico1 = df["Genero"].value_counts().plot(kind="bar",title="Grafico 1",xlabel="Genero",ylabel="Cantidad")
plt.show()

grafico2 = sns.pairplot(df[["Edad","Frecuencia"]],hue="Frecuencia",height=6)
grafico2._legend.remove()
plt.legend(title='Frecuencia',labels=df['Frecuencia'].unique().tolist())
plt.show()

grafico3 = df.plot.bar(x="Dispositivo",y="Edad",title="Grafico 3",xlabel="Dispositivo",ylabel="Edad")


```

Grafico 1: Se muestra que los encuestados prefieren mayormente las películas de acción y comedia.
Grafico 2: Se relaciona la edad con la frecuencia. A medida que aumenta la edad menor es la frecuencia en la que la gente va al cine.
Grafico 3: Se relaciona la edad con el dispositivo mayormente usado para mirar películas. La gente mayor prefiere la televisión, mientras que la gente más joven usa más los celulares y computadoras
