```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as barras2

df = pd.read_csv("/jugadores.csv",sep=",",engine="python",encoding="latin-1")
df1 = df.sort_values('Edad', ascending=False)
df1.plot.bar(x="Nombre", y="Edad", rot=70, title="Edades por club")
plt.show()

df["Club"].value_counts().plot(kind="bar")
plt.title("Cantidad de jugadores por club")
plt.xlabel("Club")
plt.ylabel("Cantidad de jugadores")
plt.show()

df["Nacionalidad"].value_counts().plot(kind="bar")
plt.title("Cantidad de jugadores por nacionalidad")
plt.xlabel("Nacionalidad")
plt.ylabel("Cantidad de jugadores")
plt.show()

df["Posicion"].value_counts().plot(kind="pie")
plt.title("Cantidad de jugadores por posicion")
plt.show()

df2 = df.sort_values('Goles', ascending=False)
df2.plot.bar(x="Nombre", y="Goles", rot=70, title="Goles por jugador")
plt.show()


```
