```python
import pandas as pd
import matplotlib.pyplot as plot
import matplotlib.pyplot as barras

df = pd.read_csv("/dat-ab-usos-2022.csv",sep=",",engine="python",encoding="latin-1")
df = df.loc[0:5000]
x_values = df["DIA"].unique()
y_values = df["DIA"].value_counts().tolist()

barras.bar(x_values,y_values)
barras.title('Usos por dia')
barras.xlabel('Dia')
barras.ylabel('Usos')
barras.show()

plot.plot(x_values,y_values);
plot.title('Usos por dia')
plot.xlabel('Dia')
plot.ylabel('Usos')
plot.show()
```

El gráfico muestra la cantiadad de usos de la sube por día. A medida que avanzan los días el uso va disminuyendo. 
Se podría haber usado la columna "Empresa", y relacionar la cantidad de veces que se viajó con cada empresa. 
Con las dos columnas se ve como a medida de que avanza el verano, el uso del colectivo baja ya que la gente está de vacaciones.
