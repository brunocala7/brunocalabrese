```python
import pandas as pd
import matplotlib.pyplot as plt1
import numpy as np


df1 = pd.read_csv("/datos.csv",sep=",")
print(f'Dataframe antes: \n {df1}')

df1.columns = df1.columns.str.replace(' ', '')

def convertirEnMinusculas (valor):
  if type(valor) == str:
    return valor.lower()
  return valor

def definirNulos (valor):
  valor = str(valor)
  if(valor == "nan"):
    valor = np.nan
  return valor

def sacarEspacios (valor):
  valor = str(valor)
  for caracter in valor: 
    if caracter.isspace():
      valor = valor.replace(caracter,'')
  return valor

def sacarLetras(valor):
  valor = str(valor)
  for caracter in valor:
    if not caracter.isnumeric() and caracter != ".":
      valor = valor.replace(caracter,'')
  return valor

def primeraLetra (valor):
  valor = str(valor)
  if len(valor) > 1:
    valor = valor[0]
  return valor

def arreglarBarra (valor): 
  valor = str(valor)
  for caracter in valor:
    if not caracter.isnumeric() and caracter != "/":
      valor = valor.replace(caracter,'/')
  return valor

def limpiarObservacion (valor): 
  valor = str(valor)
  valoresPosibles = ["hipertension","ninguna","dolorcabeza","hipertenso"]
  if valor not in valoresPosibles:
    valor = "ninguna"
  elif valor == "hipertenso":
    valor = "hipertension"
  elif valor == "dolorcabeza":
    valor = "dolor de cabeza"
  return valor

df2 = df1

df2 = df2.applymap(convertirEnMinusculas)
df2 = df2.applymap(sacarEspacios)
df2 = df2.applymap(definirNulos)
df2 = df2.dropna()
df2['LU'] = df2['LU'].apply(arreglarBarra)
df2['Altura'] = df2['Altura'].apply(sacarLetras)
df2['Peso'] = df2['Peso'].apply(sacarLetras)
df2['Sexo'] = df2['Sexo'].apply(primeraLetra)
df2['Observaci贸n'] = df2['Observaci贸n'].apply(limpiarObservacion)

a = 'm'
print(a.isnumeric())

print(f'\n \n \n Dataframe despues: \n {df2}')
```
